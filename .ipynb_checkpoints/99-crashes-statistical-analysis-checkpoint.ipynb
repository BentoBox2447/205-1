```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data/00-crash_reporting_incidents.csv', low_memory = False)
```

```python
data['First Harmful Event'].value_counts()
```

```python
data['Weather'].value_counts()
```

```python
contingency_table = pd.crosstab(
    data['ACRS Report Type'],
    data['Weather'],
    margins = True
)
contingency_table
```

```python
#Assigns the frequency values
malecount = contingency_table.iloc[0][0:13].values
femalecount = contingency_table.iloc[1][0:13].values

#Plots the bar chart
fig = plt.figure(figsize=(10, 5))
sns.set(font_scale=1.8)
categories = ["0-9","10-19","20-29","30-39","40-49","50+", 'a', 'a', 'a', 'a', 'a', 'a', 'a']
p1 = plt.bar(categories, malecount, 0.55, color='#d62728')
p2 = plt.bar(categories, femalecount, 0.55, bottom=malecount)
plt.legend((p2[0], p1[0]), ('Male', 'Female'))
plt.xlabel('Hours per Week Worked')
plt.ylabel('Count')
plt.show()
```
