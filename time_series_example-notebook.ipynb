# Time Series

What is a Time Series?

- Sequence of observations recorded at regular time intevals
- May typically be hourly, daily, weekly, monthly, quarterly, or annually (depends on the frequency of observations)
- Analyzing a time series is the preparatory step before developing a forecast of the series
- Time Series forecasting has enormous commerical potential and significance, bbecause it shows what's important to a buisness like 
    - Sales
    - Numer of visitors to asite
    - Stock price
- Time series involved understanding various aspects about the inherent nature of the series, so you can be better informed to create meaningful and accurate forcasts

# Visuializating a time series

```python
 #Time series data source: fpp pacakge in R.
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/a10.csv', parse_dates=['date'], index_col='date')

# Draw Plot
def plot_df(df, x, y, title="", xlabel='Date', ylabel='Value', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()

plot_df(df, x=df.index, y=df.value, title='Monthly anti-diabetic drug sales in Australia from 1992 to 2008.')   
```

**We can see all valuesare positive, so we can show this on both sides of the Y axis to emphasize growth**

```python
# Import Data
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/a10.csv', parse_dates=['date'], index_col='date')
df.reset_index(inplace=True)

# Prepare data
df['year'] = [d.year for d in df.date]
df['month'] = [d.strftime('%b') for d in df.date]
years = df['year'].unique()

# Prep Colors
np.random.seed(100)
mycolors = np.random.choice(list(mpl.colors.XKCD_COLORS.keys()), len(years), replace=False)

# Draw Plot
plt.figure(figsize=(16,12), dpi= 80)
for i, y in enumerate(years):
    if i > 0:        
        plt.plot('month', 'value', data=df.loc[df.year==y, :], color=mycolors[i], label=y)
        plt.text(df.loc[df.year==y, :].shape[0]-.9, df.loc[df.year==y, 'value'][-1:].values[0], y, fontsize=12, color=mycolors[i])

# Decoration
plt.gca().set(xlim=(-0.3, 11), ylim=(2, 30), ylabel='$Drug Sales$', xlabel='$Month$')
plt.yticks(fontsize=12, alpha=.7)
plt.title("Seasonal Plot of Drug Sales Time Series", fontsize=20)
plt.show()
```

**Here we broke down a time series analysis per mmonth for every year.  We can begin to see a pattern where drug sales drop every Feduary, then rising again in March, dropping in April, and ultimately continuing to rise through out the year**

**We can do a month to month boxplot to visuialize the monthy distributions of this**

```python
# Import Data
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/a10.csv', parse_dates=['date'], index_col='date')
df.reset_index(inplace=True)

# Prepare data
df['year'] = [d.year for d in df.date]
df['month'] = [d.strftime('%b') for d in df.date]
years = df['year'].unique()

# Draw Plot
fig, axes = plt.subplots(1, 2, figsize=(20,7), dpi= 80)
sns.boxplot(x='year', y='value', data=df, ax=axes[0])
sns.boxplot(x='month', y='value', data=df.loc[~df.year.isin([1991, 2008]), :])

# Set Title
axes[0].set_title('Year-wise Box Plot\n(The Trend)', fontsize=18); 
axes[1].set_title('Month-wise Box Plot\n(The Seasonality)', fontsize=18)
plt.show()
```

# Patterns in a Time Series Analysis

Time series can split into the following components
- **Trend:** A trend is an increasing or decreasing slope observed in the time series. 
- **Seasonality:** Seasonality is when there is a distinct repeated pattern observed between regular intervals, due to seasonal factors.  

**Not all time series have to have trend and seaonality.  A time series can have one or the other.**

**Cylic Behaviors:**
- Cyclic Behavior: When there is a rise and fall in pattern in the series does not happen in fixed intervals.  
- Cyclic is different than a seasonal pattern
- Cylical vs Seasonal
    - Cyclical: If the patterns are not or fixed calendar based frequencies
        - Ex: Business and other socio-economic factors
    - Seasional: Repeated factors due to seasonal factors
        - Ex: Month of year, day of month, weekdays, time of day, etc
